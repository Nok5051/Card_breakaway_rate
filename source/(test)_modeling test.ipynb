## 데이터 변환

```python
from sklearn.model_selection import train_test_split
card_data = data_test.iloc[:,2:]
card_target = data_test['Attrition_Flag']

print(card_data.shape)
print(card_target.shape)

X_train, X_test, y_train, y_test = train_test_split(card_data, card_target, 
                                                    test_size=0.2, random_state=12)
```

## XGB

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from xgboost import XGBClassifier

XGBclass = XGBClassifier()

XGBclass.fit(X_train, y_train)
```

```python
XGB = XGBclass.predict(X_test)
```

```python
print('XGB부스트 정확도 :',accuracy_score(y_test, XGB))
print('XGB부스트 F1_SCORE :',f1_score(y_test, XGB))
```
