```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings


warnings.filterwarnings(action='ignore')
%matplotlib inline
```

```python
columns_list = []

for i in range(0, 21):
    columns_list.append(i)

print(columns_list)
```

```python
data_test = pd.read_csv('../data/BankChurners.csv', usecols=columns_list)

df = data_test
```

```python
data_test['Attrition_Flag'].replace("Existing Customer", "0", inplace=True)
data_test['Attrition_Flag'].replace("Attrited Customer", "1", inplace=True)

data_test['Attrition_Flag'] = pd.to_numeric(data_test['Attrition_Flag'])
```

```python
data_target = data_test[['Attrition_Flag']]

numeric_label = ['Customer_Age','Months_on_book', 'Credit_Limit', 
                'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
               'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']

numeric_dataset = data_test[numeric_label]
```

```python
feature = numeric_dataset.columns

plt.figure(figsize=(20,60))

for i in range(len(feature)):
    plt.subplot(11,3,i+1)
    plt.title(feature[i])
    sns.kdeplot(data=data_test, x=feature[i], hue="Attrition_Flag")
plt.show()
```

# 표준화

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
Std_scaler = StandardScaler()
```

```python
Std_scaler.fit(numeric_dataset)
```

```python
std_data = Std_scaler.transform(numeric_dataset)
num_scaled = pd.DataFrame(data=std_data, columns=numeric_dataset.columns)
num_scaled
```

```python
merge_std = pd.merge(data_target, num_scaled, left_index=True, right_index=True, how='left') 
merge_std
```

```python
feature_std = merge_std.columns

plt.figure(figsize=(20,60))

for i in range(len(feature_std)-1):
    plt.subplot(11,3,i+1)
    plt.title(feature_std[i+1])
    sns.kdeplot(data=merge_std, x=feature_std[i+1], hue="Attrition_Flag")
plt.show()
```

> 로그 변환과 달리 정규분포의 첨도나 왜도가 달라지진 않는다.

# Min-Max

```python
MM_scaler = MinMaxScaler()

MM_scaler.fit(numeric_dataset)

mm_data = MM_scaler.transform(numeric_dataset)
min_max_scaled = pd.DataFrame(data=mm_data, columns=numeric_dataset.columns)
min_max_scaled
```

```python
merge_mm = pd.merge(data_target, min_max_scaled, left_index=True, right_index=True, how='left') 
merge_mm
```

```python
feature_mm = merge_mm.columns

plt.figure(figsize=(20,60))

for i in range(len(feature_mm)-1):
    plt.subplot(11,3,i+1)
    plt.title(feature_mm[i+1])
    sns.kdeplot(data=merge_mm, x=feature_mm[i+1], hue="Attrition_Flag")
plt.show()
```

### 결론 : StandardScaler 와 MinMaxScaler 를 써도 분포는 달라지지 않는다.

## VIF 지수 

```python
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor

def feature_engineering_XbyVIF(VIF_dataset):
    vif = pd.DataFrame()
    vif['VIF_Factor'] = [variance_inflation_factor(VIF_dataset.values, i)
                        for i in range(VIF_dataset.shape[1])]
    vif['Feature'] = VIF_dataset.columns
    return vif
```

```python
vif = feature_engineering_XbyVIF(df)
vif
```
