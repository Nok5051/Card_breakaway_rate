### 종속변수와 범주형, 이산형 변수 교차분석

```python
from scipy.stats import chi2_contingency

#Attrition_Flag  vs  Card_Category

card_category_crosstab=pd.crosstab(data_test['Attrition_Flag'], data_test['Card_Category'], margins=False)
 

result=chi2_contingency(observed=card_category_crosstab, correction=False)
print("1. 카이제곱 통계량:", result[0]) 
print("2. p-value:", result[1])
print("3. df:", result[2]) 
print("4. 기대값 행렬:")
pd.DataFrame(result[3]).rename(index={0:'Existing', 1:'Attrited'}, columns={0:'Blue', 1:'Gold', 2:'Silver', 3:'Platinum'})
 
```

## 4-2 VIF 
상관계수를 확인하고 어떤 변수에 다중공선성이 있는지 VIF 지수를 활용하여 확인해 봅니다.      
10 이상부터 다중공선성이 있는 것으로 간주

```python
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def feature_engineering_XbyVIF(VIF_dataset):
    vif = pd.DataFrame()
    vif['VIF_Factor'] = [variance_inflation_factor(VIF_dataset.values, i)
                        for i in range(VIF_dataset.shape[1])]
    vif['Feature'] = VIF_dataset.columns
    return vif

vif = feature_engineering_XbyVIF(data_test)
vif
```

> 다중공선성 확인 결과 가장 높은 무한대 값이 나온 Avg_Open_To_Buy 을 먼저 제거해서 다시 VIF 지수를 확인해 봅니다.

```python
data_test = data_test.drop(['Avg_Open_To_Buy'], axis=1)

vif = feature_engineering_XbyVIF(data_test)
vif
```

```python
data_test = data_test.drop(['Customer_Age'], axis=1)

vif = feature_engineering_XbyVIF(data_test)
vif
```

```python
data_test = data_test.drop(['Total_Trans_Ct'], axis=1)

vif = feature_engineering_XbyVIF(data_test)
vif
```

```python
data_test = data_test.drop(['Total_Amt_Chng_Q4_Q1'], axis=1)

vif = feature_engineering_XbyVIF(data_test)
vif
```

```python
data_test = data_test.drop(['Months_on_book'], axis=1)

vif = feature_engineering_XbyVIF(data_test)
vif
```

>최종적으로 다중공선성이 높은 5개의 변수를 제거해주었습니다.

**LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics

logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)

log = logistic_model.predict(X_test)
```

```python
print('로지스틕 회귀 정확도 :',metrics.accuracy_score(y_test, log))
print('로지스틕 회귀  정밀도 ', metrics.precision_score(y_test, log))
print('로지스틕 회귀  재현율 ', metrics.recall_score(y_test, log))
print('로지스틕 회귀  F1_SCORE :',metrics.f1_score(y_test, log))
```
